## Source: https://seaborn.pydata.org/tutorial/relational.html

```python
from sklearn import datasets
import pandas as pd
import seaborn as sns
```

```python
iris = datasets.load_iris(as_frame=True)
dir(iris)
```

```python
data = pd.concat([iris.data, iris.target], axis = 1)
```

```python
data.describe()
```

```python
data.corr()
```

```python
sns.relplot(x='sepal length (cm)', y='petal length (cm)', hue = 'target', data=data)
```

```python
sns.pairplot(data=data, hue='target')
```

```python
sns.pairplot(data=data, hue='target', diag_kind='hist')
```
